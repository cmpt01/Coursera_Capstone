```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files
from pandas import json_normalize # tranform JSON file into a pandas dataframe
import requests # library for performing HTTP requests
from bs4 import BeautifulSoup # library for xml & html scrapping
from sklearn.cluster import KMeans # import k-means from clustering stage
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import folium # map rendering library
from re import search # search html table data

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# List of postal codes of Canada starting with the letter M
url = "https://www.wien.gv.at/statistik/bevoelkerung/tabellen/bevoelkerung-bez-zr.html"
url_r = requests.get(url).text
s = requests.Session()
response = s.get(url, timeout = 10)
response
```

```python
# Process the website with beautiful soup
soup = BeautifulSoup(url_r, 'lxml')

# Retrive the Table
my_html_table = soup.find('table',{'class':'BDE-table-frame'})
```

```python
bez = []
y_2005 = []
y_2006 = []
y_2007 = []
y_2008 = []
y_2009 = []
y_2010 = []
y_2011 = []
y_2012 = []
y_2013 = []
y_2014 = []
y_2015 = []
y_2016 = []
y_2017 = []
y_2018 = []
y_2019 = []
y_2020 = []

for row in my_html_table.findAll('tr'):
        cells=row.findAll('td')
        if len(cells)==17:
#             if search('Not assigned',cells[1].find(text=True)):
#                 pass
#             else:
            bez.append(cells[0].find(text=True))
            y_2005.append(cells[1].find(text=True))
            y_2006.append(cells[2].find(text=True))
            y_2007.append(cells[3].find(text=True))
            y_2008.append(cells[4].find(text=True))
            y_2009.append(cells[5].find(text=True))
            y_2010.append(cells[6].find(text=True))
            y_2011.append(cells[7].find(text=True))
            y_2012.append(cells[8].find(text=True))
            y_2013.append(cells[9].find(text=True))
            y_2014.append(cells[10].find(text=True))
            y_2015.append(cells[11].find(text=True))
            y_2016.append(cells[12].find(text=True))
            y_2017.append(cells[13].find(text=True))
            y_2018.append(cells[14].find(text=True))
            y_2019.append(cells[15].find(text=True))
            y_2020.append(cells[16].find(text=True))
            
df = pd.DataFrame(bez, columns=['Bezirk'])
df['2005'] = y_2005
df['2006'] = y_2006
df['2007'] = y_2007
df['2008'] = y_2008
df['2009'] = y_2009
df['2010'] = y_2010
df['2011'] = y_2011
df['2012'] = y_2012
df['2013'] = y_2013
df['2014'] = y_2014
df['2015'] = y_2015
df['2016'] = y_2016
df['2017'] = y_2017
df['2018'] = y_2018
df['2019'] = y_2019
df['2020'] = y_2020

df
```
